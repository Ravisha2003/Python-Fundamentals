```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X=6*np.random.rand(100,1)-3
y=0.5*X**2+1.5*X+2+np.random.randn(100,1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X,y,color="red",label="Actual data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
plt.scatter(X,y,color="red",label="Actual Data")
plt.plot(X_test,y_pred,color="green")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#NOW APPLYING POLYNOMIAL LINEAR REGRESSION
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
reg1=LinearRegression()
```

```python
reg1.fit(X_train_poly,y_train)
y_pred1=reg1.predict(X_test_poly)
```

```python
Pscore=r2_score(y_test,y_pred1)
Pscore
```

```python
#sort X values for a smooth curve
```

```python
X_sorted=np.sort(X,axis=0)
X_sorted_poly=poly.transform(X_sorted)
y_sorted_pred=reg1.predict(X_sorted_poly)
```

```python
#plot actual data and polynomial regression curve
plt.scatter(X,y,color="red",alpha=0.6,label="Actual Data")
plt.plot(X_sorted,y_sorted_pred,color="blue",linewidth=2,label="Polynomial Fit (Degree=2)")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Polynomial Regression Fit")
plt.legend()
plt.show()
```
