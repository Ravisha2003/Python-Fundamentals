```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("Salary_Dataset.csv")
```

```python
df
```

```python
%matplotlib inline
```

```python
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.scatter(df.Experience,df.Salary,color="red",marker="*")
plt.show()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[["Experience"]],df.Salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=mx+b
9449.96232146*10+24848.203966523208
```

```python
pred=reg.predict(df[["Experience"]])
```

```python
pred
```

```python
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.scatter(df.Experience,df.Salary,color="red",marker="*")
plt.title("Experience vs Salary")
plt.plot(df.Experience,pred,color="blue",linewidth=2)
plt.show()
```
